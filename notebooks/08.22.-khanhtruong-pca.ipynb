# Theory

- $S_1$: Low speed limit (km/h)
- $S_2$: High speed limit (km/h)
- $V_s$: Variance of space


$S_{ms1} = \frac{S_1\times1000}{3600} = \frac{S_1}{3.6}$

$S_{ms2} = \frac{S_2\times1000}{3600} = \frac{S_2}{3.6}$

$\theta_1 = \arctan(\frac{T_1}{D}) = \arctan(\frac{1}{S_{ms1}}) = \arctan(\frac{3.6}{S_1})$ $\rightarrow$ Slope (angle with horizon)

$\theta_2 = \arctan(\frac{T_2}{D}) = \arctan(\frac{1}{S_{ms2}}) = \arctan(\frac{3.6}{S_2})$ $\rightarrow$ Slope (angle with horizon)

$\theta=0.5(\theta_1+\theta_2)$

$eigenvector_1 = [1, \tan(theta)]$

$eigenvector_2 = [\tan(theta), -1]$

$eigenvalue_1 = \frac{V_s}{\cos(\theta)}$

$eigenvalue_2 = \tan(\theta_1-\theta)\times eigenvalue_1$

Proof $\theta=0.5(\theta_1+\theta_2)$:

Assume $a_1 \le a_2, b_1 \le b_2$

$\sin\alpha_1=\frac{a}{h}, \cos\alpha_1=\frac{b_2-b_1}{h}=\frac{b-2b_1}{h}$

$\cos\alpha_2=\frac{b}{h}, \sin\alpha_2=\frac{a_2-a_1}{h}=\frac{a-2a_1}{h}$

$\tan\alpha = \frac{a_2}{b_2}
= \frac{h\sin(\alpha_2)+a_1}{h\cos(\alpha_1)+b_1}
= \frac{h\sin\alpha_2+\frac{a-h\sin\alpha_2}{2}}{h\cos\alpha_1+\frac{b-h\cos\alpha_1}{2}}
= \frac{2h\sin(\alpha_2)+a-h\sin\alpha_2}{2h\cos(\alpha_1)+b-h\cos\alpha_1}
= \frac{h\sin(\alpha_2)+a}{2h\cos(\alpha_1)+b}
= \frac{h\sin\alpha_2+h\sin\alpha_1}{h\cos\alpha_1+h\cos\alpha_2}
= \frac{\sin\alpha_2+\sin\alpha_1}{\cos\alpha_1+\cos\alpha_2}
= \frac{2\sin{\frac{\alpha_1+\alpha_2}{2}}\cos{\frac{\alpha_1-\alpha_2}{2}}}{2\cos{\frac{\alpha_1+\alpha_2}{2}}\cos{\frac{\alpha_1-\alpha_2}{2}}}
= \tan{\frac{\alpha_1+\alpha_2}{2}}$
